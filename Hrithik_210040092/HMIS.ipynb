```python
#Number of null values
no_null=df.isna().sum()
no_null
```

```python
column_list=df.columns
print(column_list)
```

```python
interested_columns = []

for column in column_list:
    c = column.lower()
    words = c.split()
    if not any(word in words for word in ['women', 'child', 'infant', 'infants']):
        interested_columns.append(column)
```

```python
for elements in interested_columns:
  print(elements)
print(len(interested_columns))
```

```python
interested_df = df[interested_columns]
```
